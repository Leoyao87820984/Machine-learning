```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from sklearn.cluster import KMeans
from sklearn import preprocessing,cross_validation
pd.options.display.max_columns =20
pd.options.display.max_rows = 20
PATH = 'titanic.xls '
```

### 数据预处理
1. 剔除数据，body和name明显没有区分乘客存活概率的特征，所有将其提剔除。
2. 数据量化，比如sex肯定是一个很有用的因素，但是需要将其量化。

```python
df = pd.read_excel(PATH)
df.drop(['body','name'],1,inplace=True)
df.convert_objects(convert_numeric=True)
df.fillna(0,inplace=True)
print(df.head())
def handle_non_numerical_data(df):
    columns = df.columns.values

    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1
            df[column] = list(map(convert_to_int,df[column]))
    return df
df = handle_non_numerical_data(df)
df.head()
```

```python
X = np.array(df.drop(['survived'],1).astype(float))
X = preprocessing.scale(X)   # If not doing this , i believe ticket will play a very important role. 
print(X[0].astype(float).reshape(-1,len(X[0])))  # Turn a 1D to a 2D!
y = np.array(df['survived'])

clf = KMeans(n_clusters=2)
clf.fit(X)

correct = 0
for i in range(len(X)):
    predict_me = X[i].astype(float)
    predict_me =  predict_me.reshape(-1,len(predict_me))    # -1 means let it calculate automatically 
                                                             #"Expected 2D array, got 1D array instead:"
    prediction = clf.predict(predict_me)
    if prediction[0] == y[i]:
        correct +=1

print(correct/len(X))
```
